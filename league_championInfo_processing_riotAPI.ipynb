```python
champData= pd.read_csv("LOLChampRoles.csv",error_bad_lines=False)
champData['tags']
```

```python
champData["tags"].value_counts()
```

```python
champData["tags"] = champData["tags"].apply(eval)
```

```python
champ = {}
for i in champData["tags"]:
    for j in i:
        if j not in champ:
            champ[j] = 1
        else:
            champ[j] += 1
```

```python
def to_1D(series):
 return pd.Series([x for _list in series for x in _list])
```

```python
to_1D(champData["tags"]).value_counts()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (10,10))
ax.bar(to_1D((champData["tags"])).value_counts().index,
        to_1D((champData["tags"])).value_counts().values)
ax.set_ylabel("Number of champion", size = 12)
ax.set_title("League of Legends", size = 14)
```
